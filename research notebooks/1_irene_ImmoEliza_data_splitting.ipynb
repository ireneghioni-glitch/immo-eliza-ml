### <span style="color:#c8a2c8; font-weight:bold">1 - Data splitting</span>

```python
# import libraries

import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# dataset loading
DATASET_CSV_PATH = "../data/raw/properties_final_irene.csv"

properties = pd.read_csv(DATASET_CSV_PATH)

# Bunch object - geometrically separates input and output
# target y -> 'price'
# fature X -> all the other columns except 'price_area_m2'

X, y = properties.drop(columns=['price', 'price_per_m2']), properties['price']
```

```python
# splitting dataset into train and test

# test_size: for a few tens of thousands of lines df, better between 0.2 and 0.25 (split 80/20 or 75/25)
# random_state: can be any int, by convention is 42 (keep it the same throughout the project)
X_train_raw, X_test_raw, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2)

# random_state in train_test_split controls only the mixing of rows in the df before the splitting
```

```python
# saving the shuffled rows

# X_train_raw contains 80% of rows with row data (numbers and strings)
X_train_raw.to_csv("../data/raw/train_test/X_train_raw.csv", index=False)

# X_test_raw contains 20% of rows with row data invisible to the model before final test
X_test_raw.to_csv("../data/raw/train_test/X_test_raw.csv", index=False)

# y_train contains 80% of rows with target data (price, numeric)
y_train.to_csv("../data/raw/train_test/y_train.csv", index=False)

# y_test contains 20% of rows with target data (price, numeric)
y_test.to_csv("../data/raw/train_test/y_test.csv", index=False)
```
